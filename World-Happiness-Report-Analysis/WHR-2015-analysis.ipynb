```python
df.drop(columns='Standard Error', axis = 1 , inplace=True)
```

```python
df.rename(columns={
    'Happiness Rank': 'Rank',
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)' : 'GDP',
    'Health (Life Expectancy)': 'LifeExpectancy',
    'Trust (Government Corruption)': 'Trust'
}, inplace=True) 
```

```python
df.hist(bins = 15, figsize=(12,10))
```

```python
sns.scatterplot(x='GDP', y='Score', data=df)
plt.title("GDP vs Score")
plt.show()
```

```python
sns.scatterplot(x='Freedom', y='Score', data=df)
plt.title("Freedom vs Score")
plt.show()
```

```python
sns.barplot(data = df.head(5), x = 'Country', y = 'Score', palette='Blues_d', hue='Score')
plt.title('Top 5 Countries')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
sns.barplot(data = df.tail(5), x = 'Country', y = 'Score', palette='Oranges_d', hue='Score')
plt.title('Bottom 5 Countries')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
region_mean_score = df.groupby(['Region']).agg({'Score' : 'mean'}).sort_values(by='Score')

sns.barplot(data=region_mean_score, y='Region', x='Score', palette='Pastel1', hue='Score', legend=False)
plt.xticks(rotation = 90)
plt.title('Region vice mean Score')
plt.show()
```

```python
region_mean_score = df.groupby(['Region']).agg({'Freedom' : 'mean'}).sort_values(by='Freedom')

sns.barplot(data=region_mean_score, y='Region', x='Freedom', palette='Pastel1', hue='Freedom', legend=False)
plt.xticks(rotation = 90)
plt.title('Region vice avg Freedom')
plt.show()
```

```python
df.nlargest(5, 'GDP').set_index('Country')['GDP'].plot.barh(title='Top 5 Countries by GDP')
plt.show()
```

```python
sns.barplot(data= df.nsmallest(5, 'GDP'), x = 'GDP', y='Country', hue='GDP')
plt.title('Worst Economies')
plt.show()
```

```python
sns.boxplot(data=df, x='Region', y='Score')
plt.xticks(rotation=90)
plt.title('Happiness Score by Region')
```

```python
sns.heatmap(df.drop(columns=['Country','Region','Rank']).corr(), annot=True, cmap='crest')
plt.title('Correlation among diffrent factors')
plt.show()
```

```python
# top 10 countries by Trust
top_trust = df[['Country', 'Trust', 'Score']].sort_values('Trust', ascending=False).head(10)

# Melt
top_trust_melted = top_trust.melt(id_vars='Country', value_vars=['Trust', 'Score'],
                                  var_name='Metric', value_name='Value')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=top_trust_melted, x='Country', y='Value', hue='Metric')

plt.title('Top 10 Countries by Trust vs. Their Happiness Score')
plt.xticks(rotation=45)
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

```python
top10 = df.nsmallest(10, 'Rank')
bottom10 = df.nlargest(10, 'Rank')

features = ['GDP', 'Family', 'LifeExpectancy', 'Freedom', 'Trust', 'Generosity']
top_avg = top10[features].mean()
bottom_avg = bottom10[features].mean()

compare = pd.DataFrame({'Top 10': top_avg, 'Bottom 10': bottom_avg})
compare.plot.bar(title='Top 10 vs Bottom 10 Happiness: Feature Comparison');
```

```python
country = 'India'
row = df[df['Country'] == country].iloc[0]
features = ['GDP', 'Family', 'LifeExpectancy', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual']
values = row[features].values # values of features

# Radar setup
angles = np.linspace(0, 2 * np.pi, len(features), endpoint=False).tolist()
values = np.concatenate((values, [values[0]]))
angles += [angles[0]]

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
ax.plot(angles, values, label=country)
ax.fill(angles, values, alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(features)
plt.title(f'Happiness Profile: {country}')
plt.show()
```

```python
average_score = df['Score'].mean()
high = df[df['Score'] > average_score]
low = df[df['Score'] <= average_score]

# Compare means
high_means = high.mean(numeric_only=True)
low_means = low.mean(numeric_only=True)

comparison = pd.DataFrame({'High Score Avg': high_means, 'Low Score Avg': low_means})
comparison.plot(kind='bar', figsize=(10,6), title='High vs. Low Happiness Country Averages')
```
